```python
# define map
map = folium.Map(location=[43.6532,-79.3832], zoom_start=11)

# import data frame of postcode, borough, neighbors and coordinates
data_csv = "Segmenting_Clustering_Neighborhoods_Toronto_sec2.csv"
coordinate_df = pd.read_csv(data_csv)

# iterate through the list to add to map
for location in coordinate_df.itertuples():
    label = 'Postal Code: {};  Borough: {};  Neighborhoods: {}'.format(location[1], location[2], location[3])
    label = folium.Popup(label, parse_html=True)    
    folium.CircleMarker(
        [location[-2], location[-1]],
        radius=1,
        color='red',
        fill=True,
        fill_color='#e67777',
        fill_opacity=0.7,
        parse_html=False).add_to(map) 
    folium.Circle(
        radius=500,
        popup=label,
        location=[location[-2], location[-1]],
        color='#e67777',
        fill=True,
        fill_color='#e67777'
    ).add_to(map) 
    
# Print map
map
```
